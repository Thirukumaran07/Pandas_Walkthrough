```python
import pandas as pd
df1 = pd.DataFrame(
    {
        "key": ["ко", "K1", "K2", "K3"], "Α": ["ΑΘ", "A1", "A2", "A3"], "в": ["ве", "81", "B2", "83"],
    }
)
df2 = pd.DataFrame(
    {
    "key": ["ко", "KI", "K5", "K6"], "C": ["CO", "C1", "C2", "C3"], "D": ["DO", "D1", "D2", "D3"],
    }
)
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1,df2)
```

```python
### inner join
pd.merge(df1,df2,on="key",how='inner')
```

```python
### left join
pd.merge(df1,df2,how='left',on="key")
```

```python
### right join
pd.merge(df1,df2,how='right',on="key")
```

```python
### outer join
pd.merge(df1,df2,how='outer',on="key")
```

```python
### left outer join
pd.merge(df1,df2,how='outer',on="key",indicator=True)
```
